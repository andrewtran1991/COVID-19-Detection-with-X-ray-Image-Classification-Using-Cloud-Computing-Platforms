The images are in three main folders (i.e. No_findings, Pneumonia, and Covid-19). Rename those folders and randomly assign images in each folder into a train and a test folders in the ratio 80%:20%.

```python
class_names = ['normal', 'viral', 'covid']
root_dir = 'X-Ray_Image_DataSet'
source_dirs = ['No_findings', 'Pneumonia', 'Covid-19']
```

```python
if os.path.isdir(os.path.join(root_dir, source_dirs[0])):   
    # Rename the source_dirs to the class_names
    for i, d in enumerate(source_dirs):
        os.rename(os.path.join(root_dir, d), os.path.join(root_dir, class_names[i]))
    
    # Create test folder
    os.mkdir(os.path.join(root_dir, 'test'))
    
    # Create test subfolders for each class
    for c in class_names:
        os.mkdir(os.path.join(root_dir, 'test', c))
    
    # Transfer 140 images (about 20%) of each class to the corresponding test subfolders 
    for c in class_names:
        images = [x for x in os.listdir(os.path.join(root_dir, c))]
        selected_images = random.sample(images, 140)
        for image in selected_images:
            source_path = os.path.join(root_dir, c, image)
            target_path = os.path.join(root_dir, 'test', c, image)
            shutil.move(source_path, target_path)
    
    # Create train folder
    os.mkdir(os.path.join(root_dir, 'train'))
    
    # Move the train subfolders into the train folder
    for c in class_names:
        source_path = os.path.join(root_dir, c)
        target_path = os.path.join(root_dir, 'train')
        shutil.move(source_path, target_path)
```

### Transform the images and generate dataloader objects

```python
image_transformer = transforms.Compose([
    torchvision.transforms.Resize(size=(224, 224)),  # Use this one for ResNet50
    #torchvision.transforms.Resize(size=(299, 299)), # Use this one for InceptionV3
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
os.path.join(train_dir,'normal')
```

```python
train_data = datasets.ImageFolder(train_dir, transform=image_transformer)
test_data = datasets.ImageFolder(test_dir, transform=image_transformer)
print("Length of the Training Data in Total: ")
print(len(train_data))
print("Length of the Test Data in Total: ")
print(len(test_data))
print("Details about the Training Data: ")
print(train_data)
print("Details about the Test Data: ")
print(test_data)
```

```python
batch_size = 20
num_workers=0

train_loader = torch.utils.data.DataLoader(train_data, 
                                           batch_size=batch_size, 
                                           num_workers=num_workers, 
                                           shuffle=True)

test_loader = torch.utils.data.DataLoader(test_data,
                                          batch_size=batch_size,
                                          num_workers=num_workers, 
                                          shuffle=True)
print('Number of training batches', len(train_loader))
print('Number of test batches', len(test_loader))
```

### Visualize the data

```python
class_names = ['normal', 'viral', 'covid']
def show_images(images, labels):
    plt.figure(figsize=(20, 10))
    for i, image in enumerate(images):
        plt.subplot(1, 20, i + 1, xticks=[], yticks=[])
        image = image.numpy().transpose((1, 2, 0))
        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229, 0.224, 0.225])
        image = image * std + mean
        image = np.clip(image, 0., 1.)
        plt.imshow(image)
                  
        plt.xlabel(f'{class_names[int(labels[i].numpy())-1]}')
    plt.tight_layout()
    plt.show()
```

```python
images, labels = next(iter(train_loader))
print(labels)
show_images(images, labels)
```

### ResNet50

```python
resnet = models.resnet50(pretrained=True)
```

```python
import torch.nn as nn
resn_in = resnet.fc.in_features
resnet.fc = nn.Linear(in_features=resn_in, out_features=3, bias=True)

print(resnet.fc)
if device:
    resnet.cuda()
import torch.optim as optim 
criterion = nn.CrossEntropyLoss()
#optimizer = optim.SGD(resnet.parameters(), lr=0.001)
optimizer = optim.Adam(resnet.parameters(), lr=3e-5)

```

```python
for name, child in resnet.named_children():
    print(name)
```

```python
%%time

NUM_EPOCHS = 10    
best_val_acc = 0.0
train_losses, train_accuracies = [], []
valid_losses, valid_accuracies = [], []
for epoch in range(NUM_EPOCHS):
    train_loss, train_accuracy = train(resnet, device, train_loader, criterion, optimizer, epoch)
    valid_loss, valid_accuracy, valid_results, y_true, y_pred = evaluate(resnet, device, test_loader, criterion)

    train_losses.append(train_loss)
    valid_losses.append(valid_loss)

    train_accuracies.append(train_accuracy)
    valid_accuracies.append(valid_accuracy)

    is_best = valid_accuracy > best_val_acc  # let's keep the model that has the best accuracy, but you can also use another metric.
    if is_best:
        best_val_acc = valid_accuracy
        resnet_path = os.path.join('model', 'resnet.pth')
        torch.save(resnet.state_dict(), resnet_path)
```

## Part C: Training and Deploying with the AWS SageMaker

```python
import tarfile
input_shape = [1,3,224,224]
trace = torch.jit.trace(resnet.float().eval(), torch.zeros(input_shape).float())
trace.save('model.pth')

with tarfile.open('model.tar.gz', 'w:gz') as f:
    f.add('model.pth')
```

```python
import boto3
import sagemaker
import time
from sagemaker.utils import name_from_base

role = sagemaker.get_execution_role()
sess = sagemaker.Session()
region = sess.boto_region_name
bucket = sess.default_bucket()

compilation_job_name = name_from_base('TorchVision-ResNet-Neo-2')
prefix = compilation_job_name+'/model'

model_path = sess.upload_data(path='model.tar.gz', key_prefix=prefix)

data_shape = '{"input0":[1,3,224,224]}'
target_device = 'ml_c5'
framework = 'PYTORCH'
framework_version = '1.4.0'
compiled_model_path = 's3://sagemaker-models-output'
```

```python
from sagemaker.pytorch.model import PyTorchModel
from sagemaker.predictor import Predictor

sagemaker_model = PyTorchModel(model_data=model_path,
                               predictor_cls=Predictor,
                               framework_version = framework_version,
                               role=role,
                               sagemaker_session=sess,
                               entry_point='resnet50.py',
                               source_dir='modules',
                               py_version='py3',
                               env={'MMS_DEFAULT_RESPONSE_TIMEOUT': '500'}
                              )
```

```python
compiled_model = sagemaker_model.compile(target_instance_family=target_device, 
                                         input_shape=data_shape,
                                         job_name=compilation_job_name,
                                         role=role,
                                         framework=framework.lower(),
                                         framework_version=framework_version,
                                         output_path=compiled_model_path
                                        )
```

```python
predictor = compiled_model.deploy(initial_instance_count = 1,
                                  instance_type = 'ml.c5.9xlarge'
                                 )
```
